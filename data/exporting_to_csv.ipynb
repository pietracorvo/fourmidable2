```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
#import the function from the loop library
from loop_analysis import get_averaged_loop
```

### Create the signal

```python
t = np.linspace(0, 1, 1000)
signal = np.zeros((len(t), 4))
signal[:, 0] = t
signal[:, 1] = np.sin(np.pi*2*t)

names = ['x', 'y', 'z']
for i in range(3):
    plt.plot(signal[:, 0], signal[:, i+1], label=names[i])
plt.legend()
plt.xlabel('t')
plt.ylabel('Field')
```

### Export to csv

```python
filename = 'test_signal.csv'
np.savetxt(filename, signal)
```
