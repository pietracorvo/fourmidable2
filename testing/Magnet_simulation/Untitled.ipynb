```python
%matplotlib inline
import virtual_magnet as vm
import matplotlib.pyplot as plt
import numpy as np
```

```python
# create signal and time vector
t = np.linspace(0, 3, 30000)
dt = t[1]-t[0]
signal = 0.5*np.sin(2*np.pi*t)

# get response
resp_signal = vm.magnet_response(dt, signal)

# plot
plt.plot(t, signal, label='signal')
plt.plot(t, resp_signal, label='response')
plt.grid()
```
