```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
# plt.style.use('dark_background')

# create a signal
time_l = 3
rate = 10000
t = np.arange(0, 3, 1/rate)
signal = np.zeros(t.size)
signal[np.logical_and(t<2, t>1)] = 1
plt.plot(signal)
```

```python
# run the signal through the magnet
from virtual_magnet import magnet_response
mag_signal = magnet_response(1/rate, signal)
plt.plot(mag_signal)
```

```python
mag_fft = np.fft.rfft(mag_signal)
mag_fp = pd.DataFrame(np.hstack((np.abs(mag_fft)[:, None], np.angle(mag_fft)[:, None])), columns=['amplitude', 'phase'])
sig_fft = np.fft.rfft(signal)
sig_fp = pd.DataFrame(np.hstack((np.abs(sig_fft)[:, None], np.angle(sig_fft)[:, None])), columns=['amplitude', 'phase'])
pd.merge(sig_fp, mag_fp, how='outer', left_index=True, right_index=True)
```

### Bin the data

```python
n_bins = 500
cutoff_freq = 100

# get a fourier transform of that
n = signal.size
fft = np.fft.rfft(signal)
freq = np.fft.rfftfreq(n, 1/rate)

# create DF out of fft data
signal_fft = pd.DataFrame()
signal_fft['fft_real'] = np.real(fft)
signal_fft['fft_imag'] = np.imag(fft)
signal_fft['amplitude'] = np.abs(fft)
signal_fft['phase'] = np.angle(fft)
signal_fft['freq'] = freq

# bin the data
bins = np.linspace(0, cutoff_freq, n_bins)
bins = np.append(bins, np.inf)
signal_fft['bins'] = pd.cut(signal_fft['freq'], bins=bins, include_lowest=True)

signal_fft_means = signal_fft.groupby('bins').mean().reset_index()
# signal_fft_means['amplitude'] = np.abs(signal_fft_means['fft_real']+1j*signal_fft_means['fft_imag'])
# signal_fft_means['phase'] = np.angle(signal_fft_means['fft_real']+1j*signal_fft_means['fft_imag'])
signal_fft_merged = pd.merge(signal_fft, signal_fft_means, how='left', on='bins')
signal_fft_merged.iloc[150:]
```

### Now do the reverse fourier transform on the filtered signal and run it through magnet response

```python
fft_real = np.array(signal_fft_merged['fft_real_y'].fillna(0))
fft_imag = np.array(signal_fft_merged['fft_imag_y'].fillna(0))

fft_filtered = fft_real + np.array([1j])*fft_imag

signal_filtered = np.fft.irfft(fft_filtered)
response_filtered = magnet_response(1/rate, signal_filtered)
plt.figure(figsize=(20, 6))
plt.subplot(121)
plt.plot(signal_filtered, label='filtered')
plt.plot(signal, label='true signal')
plt.legend()
plt.title('Signal')
plt.subplot(122)
plt.plot(response_filtered, label='filtered')
plt.plot(magnet_response(1/rate, signal), label='true signal')
plt.title('Magnet response')
plt.legend()
```

### Now lets see what happens if we reinstate the filtered signal by scaling to the dropped frequencies

```python
# calculate the percentages of amplitudes for each bin
def calc_bin_percentage(group_df):
    bins_sum = group_df["amplitude"].sum()
    n = group_df["amplitude"].shape[0]
    return group_df["amplitude"] / bins_sum *n

signal_perc = signal_fft.groupby("bins").apply(calc_bin_percentage).to_frame()
signal_perc.index = signal_perc.index.droplevel()
signal_perc.columns = ['percentage']
signal_fft_perc_merged = pd.merge(signal_fft_merged, signal_perc, how='left', left_index=True, right_index=True)
# signal_fft_perc_merged
# now scale the corresponding frequencies
signal_fft_perc_merged['fft_real_y'] = signal_fft_perc_merged['fft_real_y']*signal_fft_perc_merged['percentage']
signal_fft_perc_merged['fft_imag_y'] = signal_fft_perc_merged['fft_imag_y']*signal_fft_perc_merged['percentage']
signal_fft_perc_merged['amplitude_y'] = np.abs(signal_fft_perc_merged['fft_real_y']+1j*signal_fft_perc_merged['fft_imag_y'])
# signal_fft_perc_merged.loc[151:, 'amplitude_y'] *= 1562.5
signal_fft_perc_merged['phase_y'] = signal_fft_perc_merged['phase_x']
signal_fft_perc_merged['fft_real_y'] = signal_fft_perc_merged['amplitude_y']*np.cos(signal_fft_perc_merged['phase_y'])
signal_fft_perc_merged['fft_imag~_y'] = signal_fft_perc_merged['amplitude_y']*np.sin(signal_fft_perc_merged['phase_y'])
# signal_fft_perc_merged.iloc[148:]
```
