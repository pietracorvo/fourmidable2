```python
from numba import jit
@jit(cache=True)
def get_data():
    folder = "/home/luka/OneDrive/PhD/3DMOKE/Calibration data/MagCalib/"
    file_name = "MagCalib_20181120-230808.h5"
    loc = os.path.join(folder, file_name)
    file = h5py.File(loc)
    hp = np.array(file['/data8/hallprobe/data/table'])
    file.close()
#     with pd.HDFStore(loc) as store:
#         hp = store.get('/data8/hallprobe/data')
#         hx = store.get('/data8/hexapole/data')
    return pd.DataFrame(hp).set_index('index')


def get_data_nojit():
    folder = "/home/luka/OneDrive/PhD/3DMOKE/Calibration data/MagCalib/"
    file_name = "MagCalib_20181120-230808.h5"
    loc = os.path.join(folder, file_name)
    file = h5py.File(loc)
    hp = np.array(file['/data8/hallprobe/data/table'])
    file.close()
#     with pd.HDFStore(loc) as store:
#         hp = store.get('/data8/hallprobe/data')
#         hx = store.get('/data8/hexapole/data')
    return pd.DataFrame(hp).set_index('index')
     
@jit(parallel=True)
def get_data_pandas():
    int('2')
    print(os.listdir('/'))
    print(np.max([int(k.split('data')[1]) for k in ['data1', 'data123']]))
    
    tuples = [('hexapole', 'A', 'amplitude'), ('hexapole', 'A', 'phase'),
              ('hexapole', 'B', 'amplitude'), ('hexapole', 'B', 'phase'),
              ('hexapole', 'C', 'amplitude'), ('hexapole', 'C', 'phase'),
              ('hallprobe', 'A', 'amplitude'), ('hallprobe', 'A', 'phase'),
              ('hallprobe', 'B', 'amplitude'), ('hallprobe', 'B', 'phase'),
              ('hallprobe', 'C', 'amplitude'), ('hallprobe', 'C', 'phase')
              ]
    multiindx = pd.MultiIndex.from_tuples(tuples)
    folder = "/home/luka/OneDrive/PhD/3DMOKE/Calibration data/MagCalib/"
    file_name = "MagCalib_20181120-230808.h5"
    loc = os.path.join(folder, file_name)
    file = pd.HDFStore(loc)
    hp = file.get('/data8/hallprobe/data')
    file.close()
#     with pd.HDFStore(loc) as store:
#         hp = store.get('/data8/hallprobe/data')
#         hx = store.get('/data8/hexapole/data')
    return hp
```

```python
import time
t0 = time.time()
get_data()
print('jit: ', time.time()-t0)
t0 = time.time()
npd = get_data_nojit()
print('nojit: ',time.time()-t0)
t0 = time.time()
pdd = get_data_pandas()
print('pandas: ',time.time()-t0)
```
